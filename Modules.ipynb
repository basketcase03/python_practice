#### Problem2

```python
import sys
import os

dir=sys.argv[1]
file_list=os.listdir(dir)
print(file_list)

ext_dict={}

for file in file_list:
    if(len(file.split('.'))>1):
        ext=file.split('.')[1]
        ext_dict[ext] = ext_dict.get(ext, 0) + 1

for ext,count in ext_dict.items():
    print(ext, count)
```

#### Problem3

```python
import sys
import os
import datetime

dir=sys.argv[1]
file_list=os.listdir()
for file in file_list:
    print(file,os.stat(file).st_size,
          datetime.datetime.fromtimestamp(os.stat(file).st_mtime))
```

```python
import sys
import os

dir=sys.argv[1]
def get_files(dir,num):
    file_list=os.listdir(dir)
    for file in file_list:
        try:
            new_files=os.listdir(dir+"\\"+file)
            print('|','-'*2*num,file)
            get_files(dir+"\\"+file,num+2)
        except:
            print('|','-'*2*num,file)

get_files(dir,1)
        
```

#### problem7

```python
import re
```

```python
def make_slug(string):
    string= re.sub('[^a-zA-Z0-9]','-',string)
    string= re.sub('--+','-',string)
    string= re.sub('\A-','',string)
    return re.sub('-\Z','',string)
    
    
```

```python
make_slug("___;hi!--i am 2bbbb)()")
```

#### problem8

```python
from urllib.request import urlopen
import re
def get_links(url):
    request=urlopen(url)
    content=request.read().decode('utf-8')
    content=content[:]
    #print(content)
    lst=re.findall('https:[^"]*"',content)
    print(lst[0])
    print(lst[1])
    print(lst[2])
    print(lst[3])
    #for link in lst:
     #   print(link)
    print('done')
```

```python
get_links('https://en.wikipedia.org/wiki/Google')
```
